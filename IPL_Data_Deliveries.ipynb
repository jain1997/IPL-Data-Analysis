```python
def RunSummary(df_dhoni):
    print("Total Singles :{} ".format(len(df_dhoni[df_dhoni['batsman_runs']==1])*1))
    print("Total Doubles :{} ".format(len(df_dhoni[df_dhoni['batsman_runs']==2])*2))
    print("Total Triples :{} ".format(len(df_dhoni[df_dhoni['batsman_runs']==3])*3))
    print("Total Fours :{} ".format(len(df_dhoni[df_dhoni['batsman_runs']==4])*4))
    print("Total Sixes :{} ".format(len(df_dhoni[df_dhoni['batsman_runs']==6])*6))
    print("Total Run : {} ".format(df_dhoni['batsman_runs'].sum()))
```

```python
RunSummary(df_dhoni)
```

#### Replacing the team by their abbreviations

```python
df['bowling_team'].unique()
```

```python
Teams= {'Royal Challengers Bangalore':'RCB', 'Sunrisers Hyderabad':'SRH',
       'Rising Pune Supergiant':'RPS', 'Mumbai Indians':'MI',
       'Kolkata Knight Riders':'KKR', 'Gujarat Lions':'GL', 'Kings XI Punjab':'KXIP',
       'Delhi Daredevils':'DD', 'Chennai Super Kings':'CSK', 'Rajasthan Royals':'RR',
       'Deccan Chargers':'DC', 'Kochi Tuskers Kerala':'KTK', 'Pune Warriors':'PW',
       'Rising Pune Supergiants':'RPS'
    }
```

```python
df['bowling_team'] = df['bowling_team'].map(Teams)
df['batting_team'] = df['batting_team'].map(Teams)
```

```python
#Batsman Comparision
df.columns
```

```python
runs = df.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
runs
```

```python
balls = df.groupby('batsman')['ball'].count().reset_index()
```

```python
balls
```

```python
four = df[df['batsman_runs'] == 4]
```

```python
four
```

```python
df_four = four.groupby('batsman')['batsman_runs'].count().reset_index()
df_four.columns = ['batsman','4s']
df_four
```

```python
six = df[df['batsman_runs'] == 6]
six
```

```python
df_six = six.groupby('batsman')['batsman_runs'].count().reset_index()
df_six.columns=['batsman','6s']
df_six
```

```python
player=pd.concat([runs,balls.iloc[:,1],df_four.iloc[:,1],df_six.iloc[:,1]],axis=1)
player
```

```python
player['strike_rate']=(player['batsman_runs']/player['ball'])*100
player.head()
```

```python
player.dtypes
```

```python
player['4s'] = player['4s'].fillna(0)
player['6s'] = player['6s'].fillna(0)
```

```python
player['4s'] = player['4s'].astype(int)
player['6s'] = player['6s'].astype(int)
```

```python
player.head()
```

```python
#highest individual score
grp=df.groupby(['match_id','batsman','batting_team'])['batsman_runs'].sum().reset_index()
grp
```

```python
Highest = grp.groupby('batsman')['batsman_runs'].max().reset_index()
```

```python
player2=pd.concat([player,Highest.iloc[:,1:3]],axis=1)
player2
```
